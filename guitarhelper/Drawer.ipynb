```python
from guitarhelper.position import draw_chord_diagram
from io import BytesIO
import ipywidgets
from IPython.display import Image
from guitarhelper.data import guitar_chords
```

```python
collection = list(guitar_chords.values())

def draw_from_collection(i):
    chord = collection[i]
    out = BytesIO()
    draw_chord_diagram(chord["positions"], out, chord["name"])
    display(Image(out.getvalue()))
    
ipywidgets.interact(draw_from_collection, i=ipywidgets.IntSlider(min=0, max=len(collection)-1, step=1, value=0))
```
